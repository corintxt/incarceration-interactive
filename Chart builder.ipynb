```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('data/incarceration_trends.csv')
```

```python
alameda = df[df.county_name=='Alameda County']
```

```python
alameda.columns
```

```python
alt.Chart(alameda).mark_area().encode(
    x='year:O',
    y='total_jail_pretrial'
)
```

```python
from altair import Chart, X, Y, Axis, Data, DataFormat
```

```python
WIDTH = 600
HEIGHT = 300

current_county = "Alameda"
```

```python
chart = Chart(data=alameda, height=HEIGHT, width=WIDTH).mark_area(color='lightgreen').encode(
    X('year:O', axis=Axis(title='Year')),
    Y('total_jail_pretrial', axis=Axis(title='Number of inmates'))
).properties(
title='Pre-trial jail population in {}'.format(current_county)
)
```
